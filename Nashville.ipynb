```python
data_track=pd.read_csv("tracking_data.csv")
```

```python
#Convert the strings in those columns to lists because latter we will be flattening them to separate columns
data_track["homePlayers"] = data_track["homePlayers"].apply(lambda x: ast.literal_eval(x))
data_track["awayPlayers"] = data_track["awayPlayers"].apply(lambda x: ast.literal_eval(x))
data_track["ball"] = data_track["ball"].apply(lambda x: ast.literal_eval(x))
```

```python
home_pl_df = pd.DataFrame(list(itertools.chain(*data_track["homePlayers"].values.tolist())))
away_pl_df = pd.DataFrame(list(itertools.chain(*data_track["awayPlayers"].values.tolist())))
ball_df=data_track['ball'].apply(pd.Series)
```

```python
home_pl_df.columns = ['playerIdHome','numberHome','xyzHome','speedHome']
away_pl_df.columns = ['playerIdAway','numberAway','xyzAway','speedAway']
ball_df.columns=['ball_xyz','ball_speed']
```

```python
data_track = data_track.drop(['homePlayers','awayPlayers','ball'],axis=1)
```

```python
players = pd.concat([home_pl_df,away_pl_df],axis=1)
```

```python
ball_players=pd.concat([players,ball_df],axis=1)
```

```python
data_track_2 = pd.concat([data_track,ball_players],axis=1)
```

```python
data_track_2.head(10)
```

```python
data_track_2.to_csv("tracking_data_fin.csv")
```

```python
ball_coords=pd.DataFrame(data_track_2['ball_xyz'].apply(pd.Series),columns=['ball_x','ball_y','ball_z'])
```

```python
Home_coords=pd.DataFrame(data_track_2['xyzHome'].apply(pd.Series),columns=['Home_x','Home_y','Home_z'])
```

```python
Away_coords=pd.DataFrame(data_track_2['xyzAway'].apply(pd.Series),columns=['Away_x','Away_y','Away_z'])
```

```python
data_track_bh = pd.concat([ball_coords,Home_coords],axis=1)
```

```python
data_track_bha=pd.concat([data_track_bh,Away_coords],axis=1)
```

```python
data_track_fin=pd.concat([data_track_2,data_track_bha],axis=1)
```
